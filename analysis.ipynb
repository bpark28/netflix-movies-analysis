```python
import pandas as pd

# Reading csv file into dataframe
df = pd.read_csv("netflix_titles.csv")

df.head()
```

```python
# Metric 1: How fast has Netflix expanded its catalog?

# In order to find how fast catalog has expanded, 
# I need to find the number of titles added per year


# First check data type of date_added colum
df["date_added"].dtype

# Option 1
# Since data type is string, convert it to datetime
# Clean out column first (remove extra spaces in front and end)
#df["date_added"] = df["date_added"].str.strip()
#date = pd.to_datetime(df["date_added"])

# Change data type of year to int from float
#year = date.dt.year.astype('Int64')

# Option 2
# Use split() to separate years from date
split = df["date_added"].str.split(",")
year = split.str[1]

# Now, we need to find number of titles added per year
num_titles_per_year = year.value_counts()

# Sort years in chronological order
trend = num_titles_per_year.sort_index()

# Display
trend

```

```python
# Metric 2: What genres appear most frequently?

# To find what genres appear most frequently, we need to check how many genres there are and count each of them
# First check data type of listed_in column
df["listed_in"].dtype

# Use .str.split() to split each genre by "," for each index 
# and use explode() to divide up each genre in each row
split = df["listed_in"].str.split(",")
extract = split.explode()

# Clean data so there are no spaces in the front or at the end
cl_extract = extract.str.strip()

# Count each genres
genre = cl_extract.value_counts()

# Display
genre
```

```python
# Metric 3: Which countries contribute the most Netflix content?

# In order to find which countries contributed the most Netflix content,
# we need to count each countries

# First check data type of country column
df["country"].dtype

# Use .str.split() to split each country by "," if there are multiple countries
# and use explode() to divide up each countries in each row
split = df["country"].str.split(",")
extract = split.explode()

# Clean data so there are no spaces in the front or at the end
cl_extract = extract.str.strip()

# Count each countries 
countries_count = cl_extract.value_counts()

# Display
countries_count
```
